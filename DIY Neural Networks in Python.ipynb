```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time
```

```python

%matplotlib notebook
random.seed(9)

x_data = np.array([-3,-1,1,3]).reshape([-1, 1])
y_data = np.array([3,4.5,4.5,5.5]).reshape([-1, 1])

def run_model(x):
    global m,c
    y = m*x + c
    return y

m = random.uniform(-1,1)
c = 0

y_init = run_model(x_data)

fig = plt.figure()
plt.axis([-4, 4, -1, 7])
plt.scatter(x_data,y_data)
plt.plot(x_data,y_init)
fig.canvas.draw()

learn_rate = 0.01
iterations = 100
for i in range(iterations):
    for x,t in zip(x_data,y_data):
        y = run_model(x)
        error = t - y
        m = m + learn_rate*error*x
        c = c + learn_rate*error

    y_next = run_model(x_data)

    line, = plt.plot(x_data,y_next)
    fig.canvas.draw()
    line.remove()
    time.sleep(0.1)
    
plt.plot(x_data,y_next)
fig.canvas.draw()
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x_data,y_data)
y_pred = regr.predict(x_data)

plt.plot(x_data,y_pred)
fig.canvas.draw()
```
